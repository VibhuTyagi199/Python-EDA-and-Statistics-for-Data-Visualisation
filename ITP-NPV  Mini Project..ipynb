# Q-11.Using a user defined function convert the ‘emp_len’ column into
categorical column as follows -
.If emp_len is less than equals to 1 then recode as ‘fresher’.
.If emp_len is greater than 1 and less than 3 then recode as ‘junior’.
.If emp_len is greater than 3 and less than 7 then recode as ‘senior’
.If emp_len is greater than 7 then recode as ‘expert’

```python
#Answer
loan1['emp_length'].unique()
```

```python
def exp(x):
    if x in ['< 1 year','1 year']:
        return 'fresher'
    elif x in ['2 years','3 years']:
        return 'Junior'
    elif x in ['4 years', '5 years', '6 years']:
        return 'Senior'
    elif x in ['7 years','8 years','9 years','10+ years']:
        return 'Expert'
    else:
        return x
```

```python
loan1['emp_length'].apply(exp)
```

# Q-12.Find the sum of ‘loan_amnt’ for each grade and display the distribution of ‘loan_amnt’ using a pie plot.


```python
#Answer
grade_loanamnt=pd.DataFrame(loan1.groupby(loan1['grade'])['loan_amnt'].sum())
grade_loanamnt
```
